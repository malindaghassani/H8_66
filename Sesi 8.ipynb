## Waffle Option

```python
df_can
```

```python
df_dsn = df_can.loc[['Denmark','Norway','Sweden'],:]
df_dsn
```

Langkah 1 : membuat waffle untuk menentukan proporsi setiap kategori sehubungan dengan totalnya

```python
df_dsn.describe
```

Langkah 1 : untuk membuat waffle haruus menetukan proporsi setiap kategori sehubungan dengan total

```python
total_values = sum(df_dsn['Total'])
category_proportions = [(float(value)/total_values) for value in df_dsn ['Total']]

for i, proportion in enumerate (category_proportions):
    print(df_dsn.index.values[i]+':'+str(proportion))
```

```python
df_dsn['Total']
```

Langkah 2 : menentukan ukuran waffle

```python
width = 40
height =10

total_num_tiles = width * height

print('Total number of tiles is',total_num_tiles)
```

Langkah 3 : menggunakan proporsi tiap kategori untuk menentukan jumlah tiles masing-masing kategori

```python
tiles_per_category = [round(proportion * total_num_tiles)for proportion in category_proportions]

for i, tiles in enumerate (tiles_per_category):
    print(df_dsn.index.values[i]+':'+str(tiles))
```

langkah 4 : membuat matriks yang menyerupai wafel & mengisinya

```python
#1. initialize the waffle chart as empty matrix
waffle_chart = np.zeros((height,width))

#2. defne indices to loop through waffle chart
category_index = 0
tile_index = 0

#3. populate the waffle chart
for col in range (width):
    for row in range (height):
        tile_index +=  1
#4. if the no of tiles populated for current category is equal to ots corresponding allocated tiles
        if tile_index>sum(tiles_per_category[0:category_index]):
#5. proceed to next category            
            category_index +=1
#6. set the class to an integer            
        waffle_chart[row,col] = category_index
        
print ('Waffle chart populated!')
```

```python
waffle_chart
```

matriks terdiri atas 3 kategori [1,2,3] dan jumlah total instance setiap kategori sesuai dengan jumlah tiles yang dialokasikan untuk setiap katagori
