# Area Plot

```python
df_top5.head()
```

```python
df_top5.plot(kind='area',
             alpha=0.25,
             stacked = False, 
             figsize= (20,10),
            )
plt.title('Immigration Trend of Top 5 Countries')
plt.ylabel('Number of Immigrants')
plt.xlabel('Year')
plt.show()

#stacked = cumulative

df_top5.plot(kind='area',
             alpha=0.25,
             stacked = True, 
             figsize= (20,10),
            )
plt.title('Immigration Trend of Top 5 Countries')
plt.ylabel('Number of Immigrants')
plt.xlabel('Year')
plt.show()


df_top5.plot(kind='area',
             alpha=0.25, 
             figsize= (20,10),
            )
plt.title('Immigration Trend of Top 5 Countries')
plt.ylabel('Number of Immigrants')
plt.xlabel('Year')
plt.show()
```

```python
df_top5.plot(kind='area',
             alpha=0.5, 
             figsize= (20,10),
            )
plt.title('Immigration Trend of Top 5 Countries')
plt.ylabel('Number of Immigrants')
plt.xlabel('Year')
plt.show()

#Opsi 1 : scripting layer - menggunakan matplotlib.pyplot as'plt'
```

# Pie Chart

```python
df_continents = df_can.groupby('Continent',axis=0).sum()

print(type(df_can.groupby('Continent',axis=0)))

df_continents.head()
```

```python
df_continents['Total'].plot(kind='pie',
                           figsize=(5,6),
                           autopct='%1.1f%%',
                           startangle = 90,
                           shadow=True,
                           )

plt.title('Immigration to Canada by Continent [1998-2013]')

plt.show()
```

```python
colors_list = ['gold','yellowgreen','lightcoral','lightskyblue','lightgreen','pink']
explode_list =[0.1,0,0,0,0.1,0.1]

df_continents['Total'].plot(kind='pie',
                           figsize = (15,6),
                           autopct = '%1.1f%%', #ditampilkan sebagai %
                           startangle =90, #startange di 90%
                           shadow = True,
                           labels = None,
                           pctdistance =1.12, #rasio antara center of pie ke text generated
                           colors=colors_list,
                           explode=explode_list
                           )


plt.title('Immigration to Canada by Continent [1998-2013]',y=1.12)
plt.axis('equal')

plt.legend(labels=df_continents.index,loc='upper left')
plt.show()
```

```python
colors_list = ['gold','yellowgreen','lightcoral','lightskyblue','lightgreen','pink']
explode_list =[0.1,0,0,0,0.1,0.1]

df_continents['2013'].plot(kind='pie',
                           figsize = (15,6),
                           autopct = '%1.1f%%', #ditampilkan sebagai %
                           startangle =90, #startange di 90%
                           shadow = True,
                           labels = None,
                           pctdistance =1.12, #rasio antara center of pie ke text generated
                           colors=colors_list,
                           explode=explode_list
                           )


plt.title('Immigration to Canada by Continent [1998-2013]',y=1.12)
plt.axis('equal')

plt.legend(labels=df_continents.index,loc='upper left')
plt.show()
```
